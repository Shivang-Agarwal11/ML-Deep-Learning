<h1> Decision Tree </h1>

<h3>It is a tool that has applications spanning several different areas. Decision trees can be used for classification as well as regression problems. The name itself suggests that it uses a flowchart like a tree structure to show the predictions that result from a series of feature-based splits. It starts with a root node and ends with a decision made by leaves. Leaaf nodes are called as pure nodes or pure leaves.</h3>

<h3>We generally use three major concepts in Decision Tree</h3>
<ol>
    <li>Entropy</li>
    <li>Gini Impurity</li>
    <li>Information Gain</li>
</ol>


<h4> Entropy and Gini impurity are used in splitting of nodes. Entropy is considered as a method for splitting when there are generally less number of features while gini impurity is used when dataset contains many features as gini impurity calculation is faster than entropy calculation due to the presence of log in entropy.</h4>

## Implementation

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
iris=load_iris()
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(iris.data,iris.target)
```

```python
plt.figure(figsize=(15,15))
tree.plot_tree(model,filled=True)
```
